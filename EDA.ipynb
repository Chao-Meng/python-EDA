```python
# Exploratory data analysis with Pandas
# In this task you should use Pandas to answer a few questions about the Adult dataset.
# 1. How many men and women (sex feature) are represented in this dataset?
import pandas as pd
csv_file_path = '/content/adult.csv'
data = pd.read_csv(csv_file_path)
data
# sex_count = data['gender'].value_counts()
# sex_count
```

```python
# 2. What is the average age (age feature) of women?
average_age_women = data[data['gender'] == ' Female']['age'].mean()
average_age_women
# average_age_women = data[data['gender'].str.strip() == 'Female']['age'].mean()
```

```python
# 3. What is the percentage of German citizens (native-country feature)?
german_percentage = (data['native-country'] == ' Germany').mean() * 100
german_percentage
# german_percentage = (data['native-country'] == ' Germany').mean() * 100

# print(f"The percentage of German citizens is: {german_percentage:.2f}%")
```

```python
# 4. What are the mean and standard deviation of age for those who earn more than
# 50K per year (salary feature) and those who earn less than 50K per year?
# earn_more_50 = data[data['salary'] >' 50,000']['age'].mean()
age_stats_by_income = data.groupby('income')['age'].agg(['mean', 'std'])

print(age_stats_by_income)
```

```python
# 5. Is it true that people who earn more than 50K have at least a high school
# education? (education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc,
# Masters or Doctorate feature)
# 检查收入超过50K的人是否都至少具有高中教育（假设高中毕业至少需要12年的教育）
# all_have_high_school_education = all(data[data['income'] == '>50K']['educational-num'] >= 12)
# all_have_high_school_education
# 定义至少高中教育水平的学历列表
high_school_or_above = [' Bachelors', ' Prof-school', ' Assoc-acdm', ' Assoc-voc', ' Masters', ' Doctorate']

# 检查收入超过50K的人是否都至少具有高中教育
high_education_high_income = all(data[data['income'] == '>50K']['education'].isin(high_school_or_above))

high_education_high_income

```

```python
# 6. Display age statistics for each race (race feature) and each gender (sex feature).
# Use groupby() and describe(). Find the maximum age of men
# of Amer-Indian-Eskimo race.
# 显示每个种族和性别的年龄统计
age_stats_by_race_gender = data.groupby(['race', 'gender'])['age'].describe()

age_stats_by_race_gender

# 找出美洲印第安-爱斯基摩种族中男性的最大年龄
max_age_amer_indian_eskimo_male = data[(data['race'] == ' Amer-Indian-Eskimo') & (data['gender'] == ' Male')]['age'].max()
max_age_amer_indian_eskimo_male
```

```python
# 7. Among whom is the proportion of those who earn a lot (>50K) greater: married or
# single men (marital-status feature)? Consider as married those who have
# a marital-status starting with Married (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.
# 定义已婚状态
married_statuses = [' Married-civ-spouse', ' Married-spouse-absent', ' Married-AF-spouse']

# 筛选出男性
men = data[data['gender'] == ' Male']

# 计算已婚男性中收入超过50K的比例
married_men_high_income_proportion = (men[men['marital-status'].isin(married_statuses)]['income'] == '>50K').mean()

# 计算单身男性中收入超过50K的比例
single_men_high_income_proportion = (men[~men['marital-status'].isin(married_statuses)]['income'] == '>50K').mean()

print("已婚男性中收入超过50K的比例：", married_men_high_income_proportion)
print("单身男性中收入超过50K的比例：", single_men_high_income_proportion)
```

```python
# 8. What is the maximum number of hours a person works per week (hours-per-week feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?
# 找出最大的工作小时数
max_hours_per_week = data['hours-per-week'].max()

# 计算工作这么多小时的人数
people_working_max_hours = data[data['hours-per-week'] == max_hours_per_week]

# 计算在这些人中，收入超过50K的比例
proportion_high_income_max_hours = (people_working_max_hours['income'] == '>50K').mean()

print("最大的工作小时数：", max_hours_per_week)
print("工作最大小时数的人数：", len(people_working_max_hours))
print("在这些人中，收入超过50K的比例：", proportion_high_income_max_hours)
```

```python
# 9. Count the average time of work (hours-per-week) for those who earn a little and a lot (salary) for each country (native-country). What will these be for Japan?
# 根据收入水平分组，然后计算每组的平均工作小时数
average_hours_per_week_by_income = data.groupby('income')['hours-per-week'].mean()

average_hours_per_week_by_income
```

```python
# 10. Find out the total number of hours worked and mean salary as per different
# occupations.
# 计算每个职业的总工作小时数
total_hours_per_occupation = data.groupby('occupation')['hours-per-week'].sum()

# 计算每个职业中收入超过50K的比例，作为“平均收入”的近似指标
proportion_high_income_per_occupation = data.groupby('occupation').apply(lambda x: (x['income'] == '>50K').mean())

print(total_hours_per_occupation)
print(proportion_high_income_per_occupation)
```
